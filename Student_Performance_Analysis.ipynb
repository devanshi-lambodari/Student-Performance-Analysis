```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("StudentsPerformance.csv")
df.head()
```

```python
print("Dataset Info:\n")
print(df.info())

print("\nAverage Scores:\n")
print(df[['math score', 'reading score', 'writing score']].mean())

print("\nGender-wise Average:\n")
print(df.groupby('gender')[['math score', 'reading score', 'writing score']].mean())
```

DATA VISUALIZATION

```python
sns.barplot(x='gender', y='math score', data=df)
plt.title("Gender vs Math Score")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x='parental level of education', y='math score', data=df)
plt.xticks(rotation=45)
plt.title("Parental Education vs Math Score")
plt.show()
```

```python
sns.histplot(df['math score'], kde=True)
plt.title("Distribution of Math Scores")
plt.show()
```

```python
sns.heatmap(df[['math score', 'reading score', 'writing score']].corr(), annot=True)
plt.title("Correlation Between Scores")
plt.show()
```
